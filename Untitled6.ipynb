```python
# Install necessary libraries (if not installed)
!pip install matplotlib seaborn scikit-learn pandas

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage

# Load the dataset
df = pd.read_csv('/content/Indian_Vehicle_Market_Large_Dataset.csv')

# Display the first few rows of the dataset
print("Dataset Preview:")
print(df.head())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Basic statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Encoding categorical features
le = LabelEncoder()
df['Vehicle_Type_Encoded'] = le.fit_transform(df['Vehicle_Type'])
df['Fuel_Type_Encoded'] = le.fit_transform(df['Fuel_Type'])
df['Price_Range_Encoded'] = le.fit_transform(df['Price_Range'])
df['Age_Group_Encoded'] = le.fit_transform(df['Age_Group'])
df['Income_Level_Encoded'] = le.fit_transform(df['Income_Level'])
df['Geography_Encoded'] = le.fit_transform(df['Geography'])

# Correlation matrix
# Select only numerical columns for the correlation matrix
numeric_features = df.select_dtypes(include=[np.number])

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_features.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of EV Market Features')
plt.show()


# Sales Trend Over Years (2019-2022)
df['Total_Sales'] = df[['Sales_2019', 'Sales_2020', 'Sales_2021', 'Sales_2022']].sum(axis=1)
sales_years = ['Sales_2019', 'Sales_2020', 'Sales_2021', 'Sales_2022']
df[sales_years].sum().plot(kind='bar', color='skyblue', figsize=(10, 5))
plt.title('Total Sales Over the Years (2019-2022)')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()

# Distribution of Fuel Types
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Fuel_Type', palette='magma')
plt.title('Distribution of Fuel Types in the Market')
plt.show()

# EV Adoption Rate by Geography
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='Geography', y='EV_Adoption_Rate', palette='coolwarm')
plt.title('Average EV Adoption Rate by Geography')
plt.show()

# Vehicle Type vs. Price Range
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Vehicle_Type', y='Price_Range_Encoded', palette='Set3')
plt.title('Vehicle Type vs. Price Range')
plt.xlabel('Vehicle Type')
plt.ylabel('Price Range (Encoded)')
plt.show()

# Standardize the features for clustering
features = ['Sales_2019', 'Sales_2020', 'Sales_2021', 'Sales_2022',
            'EV_Adoption_Rate', 'Price_Range_Encoded', 'Age_Group_Encoded',
            'Income_Level_Encoded', 'Geography_Encoded']

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['KMeans_Segment'] = kmeans.fit_predict(df_scaled)

# Plot K-Means Clusters using PCA (for 2D visualization)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
df['PCA1'] = pca_result[:, 0]
df['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='KMeans_Segment', palette='viridis', s=60)
plt.title('K-Means Clustering Results (PCA-reduced)')
plt.show()

# Apply Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=4)
df['Agglo_Segment'] = agglo.fit_predict(df_scaled)

# Plot Dendrogram for Agglomerative Clustering
linked = linkage(df_scaled, 'ward')
plt.figure(figsize=(12, 6))
dendrogram(linked, truncate_mode='level', p=5)
plt.title('Dendrogram for Agglomerative Clustering')
plt.show()

# Analyze Segment Profiles from K-Means
# Select only numeric features for segment profiling
numeric_features = ['Sales_2019', 'Sales_2020', 'Sales_2021', 'Sales_2022',
                    'EV_Adoption_Rate', 'Price_Range_Encoded',
                    'Age_Group_Encoded', 'Income_Level_Encoded', 'Geography_Encoded']

# Group by KMeans segment and calculate the mean for each segment
segment_profiles = df.groupby('KMeans_Segment')[numeric_features].mean()

# Display the segment profiles
print("\nK-Means Segment Profiles:")
print(segment_profiles)


# Visualize Segment Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='KMeans_Segment', palette='Accent')
plt.title('Distribution of K-Means Segments')
plt.show()

# EV Adoption vs. Income Level Segmentation
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Income_Level', y='EV_Adoption_Rate', palette='cubehelix')
plt.title('EV Adoption Rate across Income Levels')
plt.show()

# Recommendations Based on Clusters
print("\nRecommendations:")
for i in range(4):
    segment_data = df[df['KMeans_Segment'] == i]
    avg_adoption = segment_data['EV_Adoption_Rate'].mean()
    common_vehicle = segment_data['Vehicle_Type'].mode()[0]
    print(f"Segment {i}:")
    print(f"  - Average EV Adoption Rate: {avg_adoption:.2f}")
    print(f"  - Most Common Vehicle Type: {common_vehicle}")
    print()

print("\nAnalysis Complete!")
```
